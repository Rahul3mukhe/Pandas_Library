# Merge

```python
import pandas as pd
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,3], 'c': [10, 11, 12]})

pd.merge(var1, var2, on='A')
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'c': [10, 11, 12]})

pd.merge(var1, var2, on='A')  #== prints only what data is common btwn both
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'c': [10, 11, 12]})

pd.merge(var1, var2,how = "left", indicator = True)
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'c': [10, 11, 12]})

pd.merge(var1, var2,how = "right", indicator = True)
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'c': [10, 11, 12]})

pd.merge(var1, var2,how = "outer", indicator = True)
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'c': [10, 11, 12]})

pd.merge(var1, var2,left_index = True, right_index = True,suffixes =("name","id"))
```

# **Concate**

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([5,6,7,8])
pd.concat([sr1,sr2])
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'B': [10, 11, 12]})
pd.concat([var1,var2],axis = 1)
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'c': [10, 11, 12]})

pd.concat([var1,var2],axis = 1, join = "inner")
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,5], 'c': [10, 11]})

pd.concat([var1,var2],axis = 1, join = "outer")
```

```python
var1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
var2 = pd.DataFrame({'A': [1,2,5], 'c': [10, 11, 12]})

pd.concat([var1,var2],axis = 1, join = "inner",keys = ["d1","d2"])
```
